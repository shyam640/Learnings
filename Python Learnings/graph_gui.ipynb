```python
# Data frame merging
import pandas as pd
frame1 = pd.DataFrame({'keys':range(5),'frame1':['a','b','c','d','e']})
frame2 = pd.DataFrame({'keys':range(2,7),'frame1':['p','q','r','s','t']})
pd.merge(frame1,frame2,on='keys')
```

```python
print(pd.merge(frame1,frame2,on='keys',how='left'),end='\n\n\n')
print(pd.merge(frame1,frame2,on='keys',how='right'),end='\n\n\n')
print(pd.merge(frame1,frame2,on='keys',how='inner'),end='\n\n\n')
print(pd.merge(frame1,frame2,on='keys',how='outer'),end='\n\n\n')
```
